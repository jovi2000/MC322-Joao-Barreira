## Desafio 1 - Empréstimo

Em um financiamento com juros compostos e número de parcelas fixas parte-se dos seguintes parâmetros:
* `S` - valor da primeira parcela
* `N` - número de parcelas
* `J` - percentual de juros mensal

A primeira parcela a ser paga do financiamento é sempre igual a `S`. A partir daí é feita uma atualização mensal da parcela, em que cada nova parcela é calculada a partir da parcela do mês anterior, conforme a fórmula:

> Parcela<sub>mês</sub> = Parcela<sub>mês-1</sub> * (1 + `J` / 100)

O financiamento encerra quando as `N` parcelas são pagas.

Exemplo:
* `S`: 200
* `N`: 5
* `J`: 1%

Parcelas do financiamento:
`200`; `202`; `204.02`; `206.06`; `208.12`

### Desafio 1 / Parte 1 - Escrevendo o programa sem módulos

Dado o problema descrito, escreva um programa que calcule as parcelas de um empréstimo para os seguintes valores:
* S: 200
* N: 5
* J: 1%

Nesta versão não use outro módulo além da função principal.

```java
double S = 200; // Valor da parcela inicial
double N = 5; // Número de parcelas 
double J = 1; // Percentual de juros mensal
double parcela_atual = S;

for (int i = 1; i <= N; i++) {
    System.out.println("Valor da parcela " + i + " é: " + parcela_atual);
    parcela_atual = parcela_atual * (1 + J/100);
}
```

### Desafio 1 / Parte 2 - Escrevendo módulos

Reescreva o código acima de forma que seu programa faça uso de uma função que seja responsável pelo cálculo de uma parcela X do empréstimo. Podem ser usadas mais funções (métodos) conforme a necessidade.

```java
/* Função que através do valor da última parcela e do percentual de juros, calcula
a próxima parcela */
public class Emprestimo {
    static double proximaParcela(double parcela_atual, double J) {
        double parcela_final;
        parcela_final = parcela_atual * (1 + J/100);
        return parcela_final;
    }
}
```

```java
double S = 200; // Valor da parcela inicial
double N = 5; // Número de parcelas 
double J = 1; // Percentual de juros mensal
double parcela_atual = S;

for (int i = 1; i <= N; i++) {
    System.out.println("Valor da parcela " + i + " é: " + parcela_atual);
    parcela_atual = Emprestimo.proximaParcela(parcela_atual, J);
}
```

## Desafio 2

Escreva um programa que sorteie um número inteiro entre `0` e `9999` o número deve ser mostrado, em seguida convertido para binário e a versão binária deve ser apresentada no console. A conversão decimal/binário deve ser computada por um programa feito por você.

```java
public class Binario {
    static void converter(int numero) {
        int quociente;
        int resto;
        String binario = ""; /* O número binário será representado por uma String
        e suas unidades vão ser adicionadas através da concatenação */
        
        do {
            quociente = numero / 2;
            resto = numero % 2;
            numero = quociente; /* Para que o método continue, precisamos dividir
            o quociente adquirido por 2, logo ele vira o número para o próximo
            loop */
            binario = binario + Integer.toString(resto); /* Pelo método que está
            sendo realizado, o resto da divisão do número por 2 é uma das unidades
            da forma binária */
        } while (quociente >= 2);
        
        binario = binario + Integer.toString(quociente); /* O último quociente
        que sobrou precisa ser adicionado na forma binária */
        
        System.out.print("Numero Sorteado na forma Binária: ");
        
        /* O método utilizado para a conversão adquire o número binário ao contrário,
        o que faz com que seja preciso percorrer e printar o número de trás para
        frente */
        for (int i = binario.length() - 1; i >= 0; i--) {
            char numero_atual = binario.charAt(i);
            System.out.print(numero_atual);
        }
        
    }
}
```

```java
import java.util.Random;

Random rand = new Random();

int n_sorteado = rand.nextInt(10000);

System.out.println("Numero Sorteado na forma Decimal: " + n_sorteado);

Binario.converter(n_sorteado);
```

## Desafio 3

Uma molécula de DNA pode ser definida a partir de uma cadeia que representa a sequência de suas bases:
A - Adenina
C - Citisina
G - Guanina
T - Tinina

Desse modo, uma string pode ser usada para representar um segmento do DNA da seguinte maneira: `ATTACGCGCAAAC`.

Escreva uma função (método) que codifique a cadeia de RNA produzida a partir de uma cadeia de DNA. A função deve ser genérica o suficiente para ser aplicável a qualquer cadeia. A entrada é uma string (DNA) e o retorno é uma string (RNA).

Escreva um programa que teste esta função com uma cadeia de DNA de sua escolha.

```java
public class Bases {
    static void codificar(String DNA) {
        String RNA = "";
        /* Percorrendo todas as bases nitrogenadas do DNA e para cada base do DNA
        uma nova base correspondente é adicionada no RNA(concatenada, já que o 
        RNA é representado por uma String) */
        for (int i = 0; i < DNA.length(); i++) {
            char base = DNA.charAt(i);
            if (base == 'A') {
                RNA = RNA + "U";
            }
            else if (base == 'C') {
                RNA = RNA + "G";
            }
            else if (base == 'G') {
                RNA = RNA + "C";
            }
            else {
                RNA = RNA + "A";
            }
        }
        System.out.println("RNA: " + RNA);
    }
}
```

```java
String DNA = "ATTACGCGCAAAC";

Bases.codificar(DNA);
```

## Desafio 4

Uma empresa precisa realizar uma estatística do salário de seus funcionários.

Para fins de teste, os salários devem ser gerados aleatoriamente com valores variando entre R\\$ 2.500 e R\\$ 15.000.

Escreva um programa que gere uma lista contendo os salários de 50 funcionários da empresa e mostre no console quantos funcionários ganham salário acima da média.

```java
import java.util.Random;

Random rand = new Random();

int salarios[] = new int[50]; /* Criando o vetor que contém o salário dos 50
funcionários */

int n_sorteado;

int media = 0, acima_da_media = 0;

/* Preenchendo o vetor de salários e calculando a média  */
for (int i = 0; i < 50; i++) {
    n_sorteado = rand.nextInt(12500) + 2500; 
    media = media + (n_sorteado / 50);
    salarios[i] = n_sorteado;
}

/* Percorrendo o vetor e verificando quantos funcionários ganham acima da média */ 
for (int i = 0; i < 50; i++) {
    if (salarios[i] > media) {
        acima_da_media += 1;
    }
}

System.out.println("O número de funcionários que ganham acima da média é: " + acima_da_media);
```
